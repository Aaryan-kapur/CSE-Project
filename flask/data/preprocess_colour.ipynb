```python
def createmaskimage(imagename,path,lst,folder):
    path = path+"/"+imagename
    if(os.path.exists(path)):
#         print(imagename+" exists")
        img = cv2.imread(path)
    
    else:
        img = np.zeros(1600*256*3)
    
        img = np.reshape(img,(256,1600,3))
    
    for i in lst:
        if(folder==1):
            img[(i-1)%256,(i-1)//256][0] += (255//2)
            img[(i-1)%256,(i-1)//256][1] += (255//2)
            img[(i-1)%256,(i-1)//256][2] += (255//2)

        if(folder==2):
            img[(i-1)%256,(i-1)//256][2]=255//2
        if(folder==3):
            img[(i-1)%256,(i-1)//256][1]=255//2
        if(folder==4):
            img[(i-1)%256,(i-1)//256][0]=255//2
        
    cv2.imwrite(path, img)
    return img
```
