```python
def meansofmeans(arr):
    ar =[]
    variance = []
    
    for i in range(10000):
        br =[]
        for j in range(50):
            br.append(arr[random.randint(0,len(arr)-1)])
        ar.append(st.mean(br))
        variance.append(st.variance(br))
#     print("Mean of means",st.mean(ar))
    return ar
```
